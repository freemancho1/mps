```python
import torch

from datetime import datetime
```

```python
print(f"PyTorch Version: {torch.__version__}")
print(f"CUDA 사용 가능 여부: {torch.cuda.is_available()}")
```

```python
if torch.cuda.is_available():
    print(f"GPU Name: {torch.cuda.get_device_name(0)}")
    print(f"CUDA Version: {torch.version.cuda}")
    print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB")

    device = torch.device("cuda")
    start = datetime.now()
    a = torch.randn(20000, 20000, device=device)
    b = torch.randn(20000, 20000, device=device)
    c = torch.matmul(a, b)
    end = datetime.now()
    print(f"MatMul Shape: {c.shape}")
    print(f"Usage GPU Memory: {torch.cuda.memory_allocated() / 1024**2:.1f} MB")
    print(f"  ⇒ GPU 연산 성공, 처리 시간: {end-start}")
else:
    print("CUDA를 사용할 수 없습니다.")
```

```python
device = torch.device("cpu")
start = datetime.now()
a = torch.randn(20000, 20000, device=device)
b = torch.randn(20000, 20000, device=device)
c = torch.matmul(a, b)
end = datetime.now()
print(f"MatMul Shape: {c.shape}")
print(f"  ⇒ CPU 연산 성공, 처리 시간: {end-start}")
```
