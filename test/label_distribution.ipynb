```python
import matplotlib.pyplot as plt 
import matplotlib.ticker as mticker 

from datetime import datetime 

from mps.pp.dataio.store import LocalParquetStore
from mps.pp.dataio.loader import HistoricalDataLoader
from mps.pp.features.labeler import TripleBarrierLabeler
from mps.core.bootstrap import bootstrap
from mps.config import cfg

_ = bootstrap()
```

```python
ticker = cfg.run.tickers[0]
```

```python
store = LocalParquetStore()
loader = HistoricalDataLoader(store)
start_date = datetime.strptime(cfg.run.start_date_str, cfg.run.date_format).date()
end_date = datetime.strptime(cfg.run.end_date_str, cfg.run.date_format).date()
load_from, bars = loader.load(ticker, start_date, end_date)
load_from, len(bars)
```

```python
labeler = TripleBarrierLabeler()
labels = labeler.label(bars)
```

```python
dist = labeler.label_distribution(labels)
total = sum(dist.values())
total
```

```python
directions = list(dist.keys())
counts = list(dist.values())
ratios = [c / total * 100 for c in counts]
ratios
```
